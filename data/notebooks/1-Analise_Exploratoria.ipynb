# 1. Análise Exploratória de Dados (AED)

O objetivo deste notebook é carregar o dataset de Churn da empresa de telecomunicações, realizar uma análise inicial para entender suas características e extrair insights visuais sobre os fatores que podem influenciar o cancelamento de clientes.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações para os gráficos
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

## 2. Carregamento e Inspeção Inicial dos Dados

```python
# O caminho do arquivo é relativo à localização do notebook
df = pd.read_csv('../data/WA_Fn-UseC_-Telco-Customer-Churn.csv')

# Exibindo as 5 primeiras linhas
df.head()
```

```python
# Verificando tipos de dados e valores nulos
df.info()
```

## 3. Análise da Variável Alvo (Churn)

Vamos investigar a proporção de clientes que cancelaram e os que não cancelaram.

```python
churn_counts = df['Churn'].value_counts()
print(f"Distribuição do Churn:\n{churn_counts}")
print(f"\nProporção do Churn:\n{df['Churn'].value_counts(normalize=True) * 100}")

# Criando o gráfico
sns.countplot(x='Churn', data=df)
plt.title('Distribuição de Clientes por Churn', fontsize=16)
plt.xlabel('Churn (Cancelamento)', fontsize=12)
plt.ylabel('Contagem de Clientes', fontsize=12)

# Salvar a imagem para o README
plt.savefig('../images/churn_distribution.png')

plt.show()
```

```python
# Usando um boxplot para ver a distribuição do tenure para cada grupo de Churn
sns.boxplot(x='Churn', y='tenure', data=df)
plt.title('Distribuição do Tempo de Contrato (Tenure) por Churn', fontsize=16)
plt.xlabel('Churn (Cancelamento)', fontsize=12)
plt.ylabel('Tempo de Contrato (Meses)', fontsize=12)
plt.show()
```
